Remove all the rows having type as 'Hex'

```python
df.drop(df.index[df['Type'] == 'Hex'], inplace=True)
df
# To reset the indices
# df = df.reset_index(drop = True)
```

Now remove the rows from the dataframe having NaN values for column 'Light'.

```python
df.dropna(subset=['Light'], inplace = True)
```

```python
len(df.Type.unique())
```

Filter out the spells on the basis of their Type and make a bar graph showing your analysis of the number of spells and their distribution in the 4 types - Charm, Jinx, Transfiguration, Curse.

```python
freq = {}
for Type in df.Type.unique():
  c = 0
  for item in df.Type :
    if item==Type:
      c+=1
    freq[Type]=c

freq

plt.figure(figsize=(10,5))
plt.bar(range(len(df.Type.unique())),list(freq.values()))  

plt.xticks(range(len(freq)),df.Type.unique())
plt.show()
```
